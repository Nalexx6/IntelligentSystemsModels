```python
import matplotlib.pyplot as plt
import numpy as np

# Variant 4
#  10.2 x1  3.2 x2 = 149
#  5.8 x1 + 16 x2 = 83
#  10.3 x1 + 7.3 x2 = 234

# Define the coefficients and constants for each equation
eq1_coeffs = [10.2, -3.2]
eq1_const = 149

eq2_coeffs = [-5.8, 16]
eq2_const = 83

eq3_coeffs = [10.3, 7.3]
eq3_const = 234

# Define the x values for each line
x_values = np.linspace(-10, 20, 100)

# Define the equations for each line
eq1 = (eq1_const - eq1_coeffs[0]*x_values) / eq1_coeffs[1]
eq2 = (eq2_const - eq2_coeffs[0]*x_values) / eq2_coeffs[1]
eq3 = (eq3_const - eq3_coeffs[0]*x_values) / eq3_coeffs[1]

# Set up the figure and axes for the plots
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, figsize=(6, 12))

# Plot the first equation on the first axis
ax1.plot(x_values, eq1, color='blue', linestyle='--')
ax1.set_ylim(-20, 20)
ax1.set_xlabel('x1')
ax1.set_ylabel('x2')
ax1.set_title('10.2x1 - 3.2x2 = 149')

# Plot the second equation on the second axis
ax2.plot(x_values, eq2, color='orange', linestyle='-.')
ax2.set_ylim(-20, 20)
ax2.set_xlabel('x1')
ax2.set_ylabel('x2')
ax2.set_title('-5.8x1 + 16x2 = 83')

# Plot the third equation on the third axis
ax3.plot(x_values, eq3, color='green', linestyle=':')
ax3.set_ylim(-20, 20)
ax3.set_xlabel('x1')
ax3.set_ylabel('x2')
ax3.set_title('10.3x1 + 7.3x2 = 234')

# Display the plots
plt.tight_layout()
plt.show()
```

```python
# Find the intersection points of all pairs of lines
coeffs = [eq1_coeffs, eq2_coeffs, eq3_coeffs]
consts = [eq1_const, eq2_const, eq3_const]
points = []
for i in range(3):
    for j in range(i+1, 3):
        A = np.array([coeffs[i], coeffs[j]])
        b = np.array([consts[i], consts[j]])
        point = np.linalg.solve(A, b)
        points.append(point)

# Set up the figure and axes for the plot
fig, ax = plt.subplots(figsize=(8, 16), dpi=100)

# Plot the three equations on the same axis
ax.plot(x_values, eq1, color='blue', linestyle='--', label='10.2x1 - 3.2x2 = 149')
ax.plot(x_values, eq2, color='orange', linestyle='-.', label='-5.8x1 + 16x2 = 83')
ax.plot(x_values, eq3, color='green', linestyle=':', label='10.3x1 + 7.3x2 = 234')

# Add the legend and grid
ax.legend()
ax.grid(color='gray', linestyle='--', linewidth=0.5)

# Add axis labels and chart name
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_title('Intersection of Three Lines')

# Set the axis limits
ax.set_xlim(-20, 20)
ax.set_ylim(-20, 20)

ax.fill([p[0] for p in points], [p[1] for p in points], color='gray', alpha=0.5)

# Add labels for the intersection points
for i, point in enumerate(points):
    label = f'P{i+1} ({point[0]:.2f}, {point[1]:.2f})'
    ax.scatter([p[0] for p in points], [p[1] for p in points], c='red')
    ax.annotate(label, point, xytext=(10, 10), textcoords='offset points', fontsize=8, color='red')

# Display the plot
plt.show()
```
